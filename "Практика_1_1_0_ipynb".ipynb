# Задание 1

**Описание:** Создайте иерархию классов для разных типов сотрудников в компании. Реализуйте родительский класс Employee и дочерние классы Manager и Developer. Каждый класс должен иметь метод для расчета зарплаты на основе различных критериев класса.


Отрабатываемый принцип: Наследование



```python
class Employee:

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary


    def salary_calculation(self):
        return self.salary


class Manager(Employee):

    def __init__(self, name, salary, bonus_rate):
        super().__init__(name, salary)
        self.bonus_rate = bonus_rate


    def salary_calculation(self):
        bonus = self.salary * (self.bonus_rate / 100)
        return self.salary + bonus


class Developer(Employee):

    def __init__(self, name, salary, number_of_projects):
        super().__init__(name, salary)
        self.number_of_projects = number_of_projects


    def salary_calculation(self):
        bonus2 = self.number_of_projects * 1000
        return self.salary + bonus2


employee_ = Employee('Екатерина', 77637)
manager_ = Manager('Александр', 75000, 9)
developer_ = Developer('Марина', 234432, 23)
print(f"Работник: {employee_.name}. "
      f"Зарплата: {employee_.salary_calculation()}")
print(f"Работник: {manager_.name}. "
      f"Зарплата: {manager_.salary_calculation()}")
print(f"Работник: {developer_.name}. "
      f"Зарплата: {developer_.salary_calculation()}")
```

# Задание 2

**Описание:** Создайте иерархию классов для различных типов транспортных средств (Необходим один родительский класс и 3 дочерних). Реализуйте метод, который позволяет каждому транспортному средству возвращать собственное описание (Метод в каждом классе должен иметь одинаковое название). Продемонстрируйте вызов данного метода для каждого транспортного средства.


Отрабатываемый принцип: Полиморфизм

```python
class Vehicle:

    def __init__(self, brand, model, year_of_release):
        self.brand = brand
        self.model = model
        self.year_of_release = year_of_release


    def characteristics(self):
        return (f"Транспортное средство: {self.brand} {self.model}, "
                f"год выпуска: {self.year_of_release}")


class Car(Vehicle):

    def __init__(self, brand, model, year_of_release, num_wheels):
        super().__init__(brand, model, year_of_release)
        self.num_wheels = num_wheels


    def characteristics(self):
        return (f"{super().characteristics()}, "
                f"количество колёс: {self.num_wheels}")

class Motorbike(Vehicle):

    def __init__(self, brand, model, year_of_release, engine_capacity):
        super().__init__(brand, model, year_of_release)
        self.engine_capacity = engine_capacity


    def description(self):
        return (f"{super().characteristics()}, "
                f"объём двигателя: {self.engine_capacity} л")


class Lorry(Vehicle):

    def __init__(self, brand, model, year_of_release, load_capacity):
        super().__init__(brand, model, year_of_release)
        self.load_capacity = load_capacity


    def description(self):
        return (f"{super().characteristics()}, "
                f"грузоподъёмность: {self.load_capacity} т")


car = Car('Maserati', 'Ghibli', 2023, 4)
motorbike = Motorbike('Suzuki', 'V-Strom DL 1000', 2013, 1000)
lorry = Lorry('Mersedes-Benz', 'Sprinter 209CDI R1', 2021, 0.78)
print(car.characteristics())
print(motorbike.characteristics())
print(lorry.characteristics())
```

# Задание 3

Онлайн-магазин:
- Создайте модель для онлайн-магазина с классами Product, Order, Customer, и ShoppingCart.
- Product включает информацию о цене, наличии на складе и категории товара.
Order обрабатывает процесс покупки, включая расчет цены с учетом скидок и налогов.
- Customer управляет информацией о пользователе и его истории заказов.
- ShoppingCart позволяет добавлять, удалять и обновлять количество товаров перед оформлением заказа.


```python
class Product:

    def __init__(self, name, price, stock, category):
        self.name = name
        self.price = price
        self.stock = stock
        self.category = category


    def is_available(self, number):
        return self.stock >= number


    def reduce_stock(self, number):
        if self.is_available(number):
            self.stock -= number
        else:
            print(f'Товар {self.name} закончился.')


class Customer:

    def __init__(self, name, telephone):
        self.name = name
        self.telephone = telephone
        self.order_history = []


    def add_to_history(self, order):
        self.order_history.append(order)


class ShoppingCart:

    def __init__(self):
        self.items = {}


    def add_product(self, product, number):
        if product.is_available(number):
            if product in self.items:
                self.items[product] += number
            else:
                self.items[product] = number
        else:
            print(f'Товар {product.name} закончился.')


    def remove_product(self, product):
        if product in self.items:
            del self.items[product]


    def update_number(self, product, number):
        if product.is_available(number):
            self.items[product] = number
        else:
            print(f'Товар {product.name} закончился.')


    def calculate_total(self, discount=0, rate=20):
        total = 0
        for product, number in self.items.items():
            total += product.price * number
        total -= total * (discount / 100)
        total += total * (rate / 100)
        return round(total)


    def clear(self):
        self.items.clear()


class Order:

    def __init__(self, customer, shopping_cart, discount=0, rate=20):
        self.customer = customer
        self.items = shopping_cart.items.copy()
        self.total_price = shopping_cart.calculate_total(discount, rate)


    def finalize_order(self):
        for product, number in self.items.items():
            product.reduce_stock(number)
        self.customer.add_to_history(self)


def main():
    products = [
        Product('Телефон', 100000, 50, 'Электроника'),
        Product('Куртка', 10000, 20, 'Одежда'),
        Product('Тетрадь', 30, 100, 'Товары для школы'),
        Product('Пылесос', 20000, 37, 'Бытовая техника'),
    ]

    name = input('Введите ваше ФИО: ')
    telephone = input('Введите ваш телефон: ')
    customer = Customer(name, telephone)

    cart = ShoppingCart()

    while True:
        print('\n1. Просмотреть товары')
        print('2. Добавить товар в корзину')
        print('3. Удалить товар из корзины')
        print('4. Просмотреть корзину')
        print('5. Оформить заказ')
        print('6. Завершение работы')

        choice = input('Выберите действие: ')

        if choice == '1':
            print('\nСписок товаров:')
            for i, product in enumerate(products, start=1):
                print(f'{i}. {product.name} - {product.price:,}₽ '
                      f'(В наличии: {product.stock})')
        elif choice == '2':
            product_id = int(input('Введите номер товара: ')) - 1
            if 0 <= product_id < len(products):
                number = int(input(f'Введите количество для '
                                   f'{products[product_id].name}: '))
                cart.add_product(products[product_id], number)
                print(f'{number} шт. {products[product_id].name} '
                      f'добавлено в корзину.')
            else:
                print('Неверный номер товара.')
        elif choice == '3':
            product_id = int(input('Введите номер для удаления: ')) - 1
            if 0 <= product_id < len(products):
                cart.remove_product(products[product_id])
                print(f'{products[product_id].name} удален из корзины.')
            else:
                print('Неверный номер товара.')
        elif choice == '4':
            print('\nКорзина:')
            if cart.items:
                for product, number in cart.items.items():
                    print(f'{product.name} - {number} шт. '
                          f'(Цена: {product.price * number:,}₽)')
                print(f'Итого: {cart.calculate_total():,}₽')
            else:
                print('Корзина пуста.')
        elif choice == '5':
            intermediate_amount = cart.calculate_total(rate=0)
            discount = 10 if intermediate_amount > 100000 else 0

            rate = 20

            print(f'Скидка: {discount}%')
            print(f'Налог: {rate}%')

            order = Order(customer, cart, discount, rate)
            order.finalize_order()

            print(f'Заказ оформлен. Сумма: {order.total_price:,}₽.')
            cart.clear()
        elif choice == '6':
            print('Работа завершена.')
            break
        else:
            print('Ошибка.')


if __name__ == '__main__':
    main()
```

# Задание 4

Симулятор космического корабля:
- Создайте симулятор управления космическим кораблем с классами SpaceShip, CrewMember, и Mission.
- SpaceShip имеет атрибуты для управления топливом, состоянием корпуса, и текущей скоростью.
- CrewMember контролирует здоровье, навыки, и роли в команде (например, пилот, инженер).
- Mission определяет цели, ресурсы, и возможные события (например, аварии, встречи с астероидами).

```python
import random


class SpaceShip:

    def __init__(self, fuel, hull_health, speed):
        self.fuel = fuel
        self.hull_health = hull_health
        self.speed = speed


    def repair_hull(self):
        self.hull_health = 100
        print('Корпус корабля отремонтирован.')


class CrewMember:

    def __init__(self, name, role, skill_level):
        self.name = name
        self.role = role
        self.skill_level = skill_level
        self.health = 100


    def task(self):
        print(f'{self.name} выполняет свою роль: {self.role}.')


class Mission:

    def __init__(self, goal, resources):
        self.goal = goal
        self.resources = resources


    def events(self, spaceship, crew):

        event = random.choice([
            'столкновение с астероидом',
            'топливный бак повреждён',
            'поломка системы жизнеобеспечения',
            'сигнал бедствия от другого корабля',
            'найдено топливо'
        ])

        print(f'Событие: {event.capitalize()}!')

        if event == 'столкновение с астероидом':
            spaceship.hull_health -= 20
            if spaceship.hull_health <= 0:
                print('Корабль разрушен! Миссия провалена.')
            else:
                print(f'Состояние корпуса после столкновения: '
                      f'{spaceship.hull_health}.')

        elif event == 'топливный бак повреждён':
            spaceship.fuel -= 20
            if spaceship.fuel < 0:
                spaceship.fuel = 0
            print(f'Утечка топлива! Осталось топлива: {spaceship.fuel}.')

        elif event == 'поломка системы жизнеобеспечения':
            for member in crew:
                member.health -= 10
                if member.health <= 0:
                    print(f'{member.name} погиб.')
                else:
                    print(f'{member.name} ранен. Здоровье: {member.health}.')

        elif event == 'сигнал бедствия от другого корабля':
            print('Экипаж принял решение помочь. Часть ресурсов потрачена.')
            self.resources -= 10
            print(f'Осталось ресурсов: {self.resources}.')

        elif event == 'удачная находка — найдено топливо':
            spaceship.fuel += 30
            print(f'Запасы топлива увеличились. '
                  f'Осталось топлива: {spaceship.fuel}.')


def main():

    ship = SpaceShip(fuel=100, hull_health=100, speed=10)

    crew = [
        CrewMember(name='Алексей', role='Пилот', skill_level=8),
        CrewMember(name='Ирина', role='Инженер', skill_level=9)
    ]

    mission = Mission(goal='Найти жизнь на планете', resources=200)

    while True:
        print('\n1. Отремонтировать корпус')
        print('2. Посмотреть экипаж')
        print('3. Случайное событие')
        print('4. Завершить игру')

        choice = input('Ваш выбор: ')

        if choice == '1':
            ship.repair_hull()
        elif choice == '2':
            for member in crew:
                member.task()
        elif choice == '3':
            mission.events(ship, crew)
        elif choice == '4':
            print('Миссия выполнена.')
            break
        else:
            print('Ошибка. Попробуйте снова.')


if __name__ == '__main__':
    main()
```
