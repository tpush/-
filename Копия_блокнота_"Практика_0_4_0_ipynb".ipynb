**Задача 1:**



Интернет-магазин предлагает следующие условия скидок:

*   Для заказов больше 1000 единиц, клиент получает скидку 5%. Если клиент использует промокод SUPERDISCOUNT, он получает скидку 10%.
*  Для заказов стоимостью более 5000 единиц, клиент получает скидку 15%, а использование промокода SUPERDISCOUNT увеличивает скидку до 20%.

Этап 1:
Ввод:
```
Введите стоимость единицы товара: 5
Введите количество товара: 1000
Введите промокод: GiVEMEDISCONT
```

Вывод:

```
Ваша скидка: 5%
Итоговая сумма: 4750.0
```
Этап 2:

Оформите ваш код в виде функции

````python
def discount(a, b, c):
    d = a * b
    e = 0
    if d > 5000:
        e = 15
        if c == 'SUPERDISCOUNT':
            e = 20
    elif d > 1000:
        e = 5
        if c == 'SUPERDISCOUNT':
            e = 10
    f = d * (e / 100)
    g = d - f
    return e, g
a = int(input('Введите стоимость единицы товара: '))
b = int(input('Введите количество товара: '))
c = input('Введите промокод: ')
e, g = discount(a, b, c)
print('Ваша скидка: ' + str(e) + '%')
print('Итоговая сумма: ' + str(g))
````

**Задача 3:**

Этап 1:
Напишите программу реализующую Алгоритм Евклида


> Алгоритм Евклида – это алгоритм нахождения наибольшего общего делителя (НОД) пары целых чисел.



Ввод:
```
30 18
```

Вывод:

```
6
```

Этап 2:
Оформите ваш код в виде функции


````python
def evklid(a, b):
    while a != 0 and b != 0:
        if a > b:
            a = a % b
        else:
            b = b % a
    if a != 0:
        return a
    else:
        return b
a, b = map(int, input().split())
print(evklid(a, b))
````

**Задача 4:**

Этап 1:
Напишите функцию программу, которая принимает строку и возвращает список слов и количество их упомнинаний в предложении

Этап 2:
Оформите ваш код в виде функции

Ввод:
```
apple banana apple
```

Вывод:

```
apple: 2,
banana: 1
```

````python
def upomin (a):
    b = a.split()
    c = {}
    for i in b:
        if i in c:
            c[i] += 1
        else:
            c[i] = 1
    d = []
    for i in c:
        d.append((i, c[i]))
    return d
a = input()
print(*upomin(a))
````

**Задача 5:**

Этап 1:
Детектор анаграмм Напишите программу на Python, которая принимает в качестве входных данных две строки и проверяет, являются ли они анаграммами друг друга

Этап 2:
Оформите ваш код в виде функции

Ввод:
```
listen, silent
```

Вывод:

```
True
```

````python
def anagram(a, b):
    if len(a) != len(b):
        return False
    a1 = list(a)
    a2 = list(b)
    cnt = 0
    for i in a1:
        if i in a2:
            a2.remove(i)
            cnt += 1
    if cnt == len(a1):
        return True
    else:
        return False
gr = input('Введите слова через пробел: ')
a, b = gr.split()
print(anagram(a, b))
````

Задача: «Банковская система»

Создайте программу Python, которая имитирует базовую банковскую систему. Система должна иметь следующие функции:

Требования
*   Система должна позволять клиентам создавать счета и хранить их балансы.
*   Система должна позволять клиентам вносить и снимать деньги со своих счетов.
*   Система должна позволять клиентам проверять свой текущий баланс.
*   Система должна позволять клиентам переводить деньги между счетами.
*   Система должна отслеживать транзакции (депозиты, снятия и переводы) и иметь возможность печатать детали транзакций.


Задачи
1. Реализуйте банковскую систему, используя только базовые конструкции Python, такие как def, lists, if, elif и else, без классов или словарей.
Определите функции для создания счетов, внесения и снятия денег, получения балансов счетов, перевода денег между счетами, а также создания и печати транзакций.
2. Напишите основную функцию, которая демонстрирует использование банковской системы путем создания счетов, внесения и снятия денег и перевода денег между счетами.
3. Бонусное задание
Реализуйте способ хранения и печати истории транзакций для каждого счета.

Ограничения
Не используйте классы или словари.
Используйте только базовые конструкции Python, такие как def, lists, if, elif и else.


````python
acc = []
ostatok = []

def sozdat_acc():
    nameofacc = input('Функция-создание аккаунта. Введите имя: ')
    acc.append(nameofacc)
    ostatok.append(0)
    print('Счёт создан. Имя счёта: ' + nameofacc)

def vnos():
    nameofacc = input('Функция-внесение денежных средств. Введите имя: ')
    if nameofacc in acc:
        summ = float(input('Сумма: '))
        i = acc.index(nameofacc)
        ostatok[i] += summ
        print('На счёт ' + nameofacc + ' внесено ' + str(summ) + '. Остаток: ' + str(ostatok[i]))
    else:
        print('Попробуте ввести имя счёта ещё раз.')

def snatie():
    nameofacc = input('Функция-снятие денежных средств. Введите имя: ')
    if nameofacc in acc:
        summ = float(input('Сумма: '))
        i = acc.index(nameofacc)
        if ostatok[i] >= summ:
            ostatok[i] -= summ
            print('С счёта ' + nameofacc + ' снято ' + str(summ) + '. Остаток: ' + str(ostatok[i]))
        else:
            print('Недостаточно денежных средств')
    else:
        print('Попробуте ввести имя счёта ещё раз.')

def proverka_ostatok():
    nameofacc = input('Функция-проверка остатка денежных средств. Введите имя: ')
    if nameofacc in acc:
        i = acc.index(nameofacc)
        print('Счёт: ' + nameofacc + '. Остаток: ' + str(ostatok[i]))
    else:
        print('Попробуте ввести имя счёта ещё раз.')

def perevod():
    acc_for_perevod = input('Функция-перевод денег. Введите имя счёта, с которого хотите перевести деньги: ')
    if acc_for_perevod in acc:
        acc_for_perevod2 = input('Введите имя счета, на который хотите перевести деньги: ')
        if acc_for_perevod2 in acc:
            summ = float(input('Сумма: '))
            from_i = acc.index(acc_for_perevod)
            to_i = acc.index(acc_for_perevod2)
            if ostatok[from_i] >= summ:
                ostatok[from_i] -= summ
                ostatok[to_i] += summ
                print('Отправитель: ' + acc_for_perevod + '. Получатель: ' + acc_for_perevod2)
                print('Остаток отправителя (' + acc_for_perevod + '): ' + str(ostatok[from_i]))
                print('Остаток получателя (' + acc_for_perevod2 + '): ' + str(ostatok[to_i]))
            else:
                print('Недостаточно денежных средств.')
        else:
            print('Попробуйте ещё раз ввести имя счёта.')
    else:
        print('Попробуйте ещё раз ввести имя счёта')

def main():
    while True:
        print('1. Создание счёта')
        print('2. Внесение денежных средств')
        print('3. Снятие денежных средств')
        print('4. Проверка остатка')
        print('5. Перевод денежных средств с одного счёта на другой')
        print('6. Завершение работы с системой')
        vybor = input('Введите номер команды: ')
        if vybor == '1':
            sozdat_acc()
        elif vybor == '2':
            vnos()
        elif vybor == '3':
            snatie()
        elif vybor == '4':
            proverka_ostatok()
        elif vybor == '5':
            perevod()
        elif vybor == '6':
            print('Работа завершена.')
            break
        else:
            print('Ошибка. Попробуйте ввести команду ещё раз!')
if __name__ == '__main__':
    main()
````
