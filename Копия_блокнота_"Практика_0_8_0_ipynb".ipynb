```python
import requests


API_key = '97gRyCa7ME28M09YdaT7KLXaxMY5Oprhp2ylX6PZ'


def apod():  #команда №1: Астрономическая картинка дня (APOD)
    url = f'https://api.nasa.gov/planetary/apod?api_key={API_key}'
    response = requests.get(url)
    if response.status_code == 200:
        json = response.json()
        print(f"\nНазвание: {json['title']}")
        print(f"Описание: {json['explanation']}")
        print(f"Ссылка на картинку: {json['url']}\n")
    else:
        print('\nПроизошла ошибка при подключении к сайту.\n')


def mars_rover_photos(sol, camera):  #команда №2: Фотографии с марсохода
    url = f'https://api.nasa.gov/mars-photos/api/v1/rovers/curiosity/photos?sol={sol}&camera={camera}&api_key={API_key}'
    response = requests.get(url)
    if response.status_code == 200:
        json = response.json()
        if json['photos']:
            for photo in json['photos']:
                print(f"\nСсылка на фото: {photo['img_src']}")
        else:
            print('\nФотографии по заданным параметрам не найдены.\n')
    else:
        print('\nПроизошла ошибка при подключении к сайту.\n')


def asteroids_neows(date): #команда №3: Объекты, сближающиеся с Землёй (ОСЗ)
    url = (f'https://api.nasa.gov/neo/rest/v1/feed?start_date={date}'
           f'&end_date={date}&api_key={API_key}')
    response = requests.get(url)
    if response.status_code == 200:
        json = response.json()
        if date in json['near_earth_objects']:
            print(f"\nОбъекты, сближающиеся с Землёй на дату {date}:")
            for asteroid in json['near_earth_objects'][date]:
                diameter = asteroid['estimated_diameter']['meters']
                min_d, max_d = (diameter['estimated_diameter_min'],
                                diameter['estimated_diameter_max'])
                hazardous = 'Да' \
                    if asteroid['is_potentially_hazardous_asteroid'] \
                    else 'Нет'
                distance = \
                    (asteroid)['close_approach_data'][0]['miss_distance']['kilometers']

                print(f"\nНазвание: {asteroid['name']}")
                print(f"Размер (м): от {min_d} до {max_d}")
                print(f"Опасность: {hazardous}")
                print(f"Расстояние от Земли (км): {distance}")
        else:
            print("\nДанных об объектах на указанную дату нет.\n")
    else:
        print('\nОшибка при подключении к сайту, попробуйте позже\n')


def donki_FLR(startDate, endDate):  #команда №4 (1): Данные о солнечных вспышках
    url = f'https://api.nasa.gov/DONKI/FLR?startDate={startDate}&endDate={endDate}&api_key={API_key}'
    response = requests.get(url)
    if response.status_code == 200:
        json = response.json()
        if json:
            print('\nДанные о солнечных вспышках за указанный период:')
            for flare in json:
                print(f"\nДата вспышки: {flare['beginTime']}")
                print(f"Класс вспышки: {flare['classType']}")
                print(f"Источник: {flare['sourceLocation']}")
        else:
            print('\nДанных о вспышках за указанный период нет.\n')
    else:
        print('\nПроизошла ошибка при подключении к сайту.\n')


def donki_GST(startDate, endDate):  #команда №4 (2): Данные о геомагнитных бурях
    url = f'https://api.nasa.gov/DONKI/GST?startDate={startDate}&endDate={endDate}&api_key={API_key}'
    response = requests.get(url)
    if response.status_code == 200:
        json = response.json()
        if json:
            print('\nДанные о геомагнитных бурях за указанный период:')
            for storm in json:
                print(f"\nДата начала бури: {storm['startTime']}")
                print(f"Все активности: {storm['allKpIndex']}")
        else:
            print('\nДанных о бурях за указанный период нет.\n')
    else:
        print('\nПроизошла ошибка при подключении к сайту.\n')


def date_check(date_):
    return (len(date_) == 10 and date_[4] == '-' and
            date_[7] == '-' and date_[:4].isdigit() and
            date_[5:7].isdigit() and date_[8:].isdigit())


while True:
    print('\n1. Астрономическая картинка дня (APOD)')
    print('2. Фотографии с марсохода')
    print('3. Объекты, сближающиеся с Землёй (ОСЗ)')
    print('4. Данные о космической погоде')
    print('5. Завершение работы')
    a = input('Введите номер команды: ')

    if a == '1':
        apod()
    elif a == '2':
        sol = input('Введите сол: ')
        camera = input('Введите камеру: ').upper()
        mars_rover_photos(sol, camera)
    elif a == '3':
        date = input('Введите дату (формат: ГГГГ-ММ-ДД): ')
        if date_check(date):
            asteroids_neows(date)
        else:
            print('Ошибка: неверный формат даты. Формат: ГГГГ-ММ-ДД.')
    elif a == '4':
        startDate = input('Введите начальную дату (формат: ГГГГ-ММ-ДД): ')
        endDate = input('Введите конечную дату (формат: ГГГГ-ММ-ДД): ')
        if date_check(startDate) and date_check(endDate):
            print('1. Солнечные вспышки (FLR)')
            print('2. Геомагнитная буря (GST)')
            choice = input('Выберите событие: ')
            if choice == '1':
                donki_FLR(startDate, endDate)
            elif choice == '2':
                donki_GST(startDate, endDate)
            else:
                print('Ошибка. Неправильный ввод. Введите ещё раз')
        else:
            print('Ошибка. Пожалуйста, используйте формат: ГГГГ-ММ-ДД.')
    elif a == '5':
        print('Работа завершена')
        break
    else:
        print('Ошибка. Неправильный ввод. Введите ещё раз')
```

Цель этой задачи - создать скрипт на Python, который взаимодействует с API Чикагского института искусств (https://api.artic.edu/docs/) для извлечения и отображения произведений искусства. Скрипт должен позволять пользователям просматривать работы по страницам, фильтровать их по имени художника и просматривать подробную информацию о выбранных произведениях искусства. Ниже приведены требования и функциональные возможности, которые необходимо реализовать:

Требования:
Извлекать произведения искусства:

- Создайте функцию, которая извлекает список произведений искусства из API Чикагского института искусств.
Функция должна принимать параметр page для разбивки на страницы и возвращать список произведений искусства вместе с информацией о разбивке на страницы.
Фильтровать произведения искусства:

- Реализуйте функцию, которая фильтрует список произведений искусства на основе имени указанного художника. Функция должна возвращать список работ, которые соответствуют имени художника (без учета регистра).
Отображать подробную информацию об оформлении:

- Напишите функцию, которая отображает названия работ для пользователя и позволяет ему выбрать одну из них, введя соответствующий номер.
После выбора функция должна отображать подробную информацию о выбранном произведении, включая название, исполнителя, дату и носитель.
Разбивка на страницы и взаимодействие с пользователем:

- Создайте основную функцию, которая управляет выборкой произведений и взаимодействием с пользователем.

Разрешите пользователям перемещаться по страницам с произведениями искусства, выполнять фильтрацию по исполнителю или выходить из программы.

Если страниц с произведениями искусства несколько, укажите варианты перехода к следующей странице, предыдущей странице, фильтрации по исполнителю или выхода из программы.

```python
import requests


url = 'https://api.artic.edu/api/v1/artworks'


def fetch_artworks(page=1):
    params = {
        'page': page,
        'limit': 10,
        'fields': 'id,title,artist_display,date_display,medium_display'
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        artworks = data.get('data', [])
        pagination = data.get('pagination', {})
        return artworks, pagination


def filter_artworks(artworks, artist_name):
    return [
        art for art in artworks
        if 'artist_display' in art and
           artist_name.lower() in art['artist_display'].lower()
    ]


def display_details(artwork):
    print('\n- - Детали - -')
    title = artwork.get('title', 'Неизвестно')
    artist_display = artwork.get('artist_display', 'Неизвестен')
    date_display = artwork.get('date_display', 'Неизвестно')
    medium_display = artwork.get('medium_display', 'Неизвестно')
    print('Название:', title)
    print('Исполнитель:', artist_display)
    print('Дата:', date_display)
    print('Носитель:', medium_display)
    print('- - - - - - - - -')


def main():
    page = 1
    while True:
        artworks, pagination = fetch_artworks(page)
        if not artworks:
            print('Нет данных')
            break

        print(f'\n- - Страница {page} - -')
        for index, art in enumerate(artworks, start=1):
            title = art.get('title', 'Без названия')
            artist_display = art.get('artist_display', 'Неизвестен')
            print(f"{index}. {title} - {artist_display}")

        print('\nВыберите действие:')
        print('1. Следующая страница')
        print('2. Предыдущая страница')
        print('3. Фильтрация по имени художника')
        print('4. Выбрать произведение для просмотра деталей')
        print('5. Завершение работы')

        action = input('\nВыберите действие:\n').strip().lower()

        if action == '1':
            if (pagination.get('current_page', 1) <
                    pagination.get('total_pages', 1)):
                page += 1
            else:
                print('Это последняя страница')
        elif action == '2':
            if page > 1:
                page -= 1
            else:
                print('Это первая страница')
        elif action == '3':
            artist_name = input('Введите имя художника: ').strip()
            filtered_artworks = filter_artworks(artworks, artist_name)
            if filtered_artworks:
                for index, art in enumerate(filtered_artworks, start=1):
                    title = art.get('title', 'Без названия')
                    artist_display = art.get('artist_display', 'Неизвестен')
                    print(f"{index}. {title} - {artist_display}")
            else:
                print('Работы не найдены')
        elif action == '4':
            detail = int(input('Введите номер произведения: '))
            if 1 <= detail <= len(artworks):
                display_details(artworks[detail - 1])
            else:
                print('Некорректный номер')
        elif action == '5':
            print('Работа завершена')
            break
        else:
            print('Некорректный ввод')


if __name__ == '__main__':
    main()
```

# Задание 3

Задача: Создать программу по управлению портфелем криптовалют

Цель: Создать скрипт на Python, который извлекает цены на криптовалюты в режиме реального времени, позволяет пользователям управлять портфелем криптовалют, вычисляет общую стоимость портфеля, отслеживает изменения цен и предоставляет исторические данные о ценах для анализа.

Требования:
Получение текущих цен на криптовалюты:

Используйте https://docs.coingecko.com/ для получения актуальных цен на список криптовалют.

Управление портфелем:

- Позволяет пользователю создавать портфель криптовалют и управлять им, указывая количество каждой криптовалюты, которой он владеет.
- Расчитывает общую стоимость портфеля в указанной фиатной валюте (например, долларах США).

Отслеживание изменения цен:

- Отображение процентного изменения цены для каждой криптовалюты в портфеле за последние 24 часа.
- Выделите все криптовалюты, стоимость которых значительно увеличилась или снизилась.

Поиск исторических данных о ценах:

- Получение исторических данных о ценах на указанную криптовалюту за последнюю неделю.
- Предоставьте пользователю возможность визуализировать эти данные в простом текстовом формате (например, цены за день).

Взаимодействие с пользователем:

- Реализуйте интерфейс командной строки для ввода данных пользователем.
- Предоставьте опции для получения текущих цен, управления портфелем, просмотра изменений цен или анализа исторических данных.


```python
import requests
from datetime import datetime


portfolio = {}
url = "https://api.coingecko.com/api/v3"


def get_price(coin_id, currency):
    response = requests.get(f"{url}/simple/price", params={
        "ids": coin_id,
        "vs_currencies": currency,
        "include_24hr_change": "true"
    })
    json = response.json()
    if coin_id in json and currency in json[coin_id]:
        return (json[coin_id][currency],
                json[coin_id].get(f'{currency}_24h_change', 0))


def add_(coin_id, amount):
    portfolio[coin_id] = portfolio.get(coin_id, 0) + amount
    print(f'{amount} {coin_id} добавлено в портфель.')


def remove_from_portfolio(coin_id, amount):
    if coin_id in portfolio:
        portfolio[coin_id] -= amount
        if portfolio[coin_id] <= 0:
            del portfolio[coin_id]
        print(f'{amount} {coin_id} удалено из портфеля.')
    else:
        print('Ошибка')


def get_portfolio_information(currency):
    total_value = 0
    significant_changes = []
    for coin_id, amount in portfolio.items():
        price, change = get_price(coin_id, currency)
        if price > 0:
            total_value += price * amount
            print(f"{coin_id}: {amount} * {price} = {amount * price} "
                  f"{currency} (изменение за 24ч: {change}%)")
            if abs(change) >= 5:
                significant_changes.append((coin_id, change))
    print(f'Общая стоимость портфеля: {total_value} {currency}')
    if significant_changes:
        print('\nКриптовалюты с существенными изменениями за 24 часа:')
        for coin_id, change in significant_changes:
            print(f'{coin_id}: {change}%')


def get_historical_data(coin_id, currency):
    response = requests.get(f'{url}/coins/{coin_id}/market_chart', params={
        'vs_currency': currency,
        'days': '7'
    })
    json = response.json()
    if "prices" in json:
        print(f"\nИсторические данные для {coin_id} за последние 7 дней:")
        for timestamp_ms, price in json["prices"]:
            # Конвертация временной метки из миллисекунд в формат даты
            date = (datetime.fromtimestamp(timestamp_ms / 1000).
                    strftime('%d.%m.%Y %H:%M:%S'))
            print(f"Дата: {date} - Цена: {price} {currency}")


def main():
    currency = 'usd'
    while True:
        print('\n1. Просмотр текущих цен')
        print('2. Управление портфелем')
        print('3. Просмотр стоимости портфеля')
        print('4. Исторические данные')
        print('5. Изменить фиатную валюту')
        print('6. Выйти')
        choice = input('Выберите действие: ')
        if choice == '1':
            coin_id = input('Введите ID криптовалюты: ').strip()
            price, change = get_price(coin_id, currency)
            if price > 0:
                print(f'Текущая цена {coin_id}: {price} {currency} '
                      f'(изменение за 24ч: {change}%)')
            else:
                print('Ошибка')
        elif choice == '2':
            action = input('1 - добавить / 2 - удалить: ').strip().lower()
            coin_id = input('Введите ID криптовалюты: ').strip()
            amount = float(input('Введите количество: '))
            if action == '1':
                add_(coin_id, amount)
            elif action == '2':
                remove_from_portfolio(coin_id, amount)
            else:
                print('Ошибка')
        elif choice == '3':
            get_portfolio_information(currency)
        elif choice == '4':
            coin_id = input('Введите ID криптовалюты: ').strip()
            get_historical_data(coin_id, currency)
        elif choice == '5':
            new_currency = input('Код фиатной валюты: ').strip().lower()
            currency = new_currency
            print(f'Вы выбрали валюту: {currency.upper()}')
        elif choice == '6':
            print('Работа завершена')
            break
        else:
            print('Ошибка')


if __name__ == '__main__':
    main()
```
